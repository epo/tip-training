
# PATSTAT Tables and Fields

## Application Tables

### 1. `TLS201_APPLN`
- **Fields:**
  - `appln_id`: Application identification number
  - `appln_auth`: Authority responsible for the application
  - `appln_nr`: Application number
  - `appln_kind`: Kind of application
  - `appln_filing_date`: Filing date of the application
  - `ipr_type`: Type of IP rights
  - `grant_year`: Year of grant (if applicable)
  - `granted`: Whether the application is granted

### 2. `TLS202_APPLN_TITLE`
- **Fields:**
  - `appln_id`: Application identification number
  - `appln_title`: Title of the application
  - `title_lg`: Language of the title

### 3. `TLS203_APPLN_ABSTR`
- **Fields:**
  - `appln_id`: Application identification number
  - `appln_abstract`: Abstract of the application
  - `abstract_lg`: Language of the abstract

### 4. `TLS204_APPLN_PRIOR`
- **Fields:**
  - `appln_id`: Application identification number
  - `prior_appln_id`: Prior application identification number
  - `prior_appln_seq_nr`: Sequence number of the prior application

### 5. `TLS205_TECH_REL`
- **Fields:**
  - `appln_id`: Application identification number
  - `tech_rel_appln_id`: Technologically related application ID
  - `rel_type`: Type of technological relationship

### 6. `TLS207_PERS_APPLN`
- **Fields:**
  - `person_id`: Person identification number
  - `appln_id`: Application identification number
  - `applt_seq_nr`: Applicant sequence number
  - `invt_seq_nr`: Inventor sequence number

### 7. `TLS209_APPLN_IPC`
- **Fields:**
  - `appln_id`: Application identification number
  - `ipc_class_symbol`: IPC classification symbol

### 8. `TLS210_APPLN_NACE`
- **Fields:**
  - `appln_id`: Application identification number
  - `nace_code`: NACE code for the application

### 9. `TLS211_PAT_PUBLN`
- **Fields:**
  - `pat_publn_id`: Patent publication identification number
  - `appln_id`: Application identification number
  - `publn_auth`: Publication authority
  - `publn_nr`: Publication number
  - `publn_kind`: Kind of publication
  - `publn_date`: Publication date

### 10. `TLS212_CITATION`
- **Fields:**
  - `citn_id`: Citation identification number
  - `appln_id`: Application identification number
  - `citn_origin`: Origin of the citation
  - `citn_seq_nr`: Sequence number of the citation

### 11. `TLS214_NPL_PUBLN`
- **Fields:**
  - `npl_publn_id`: Non-patent literature publication ID
  - `npl_biblio`: Bibliographic information of non-patent literature

### 12. `TLS215_CITN_CATEG`
- **Fields:**
  - `citn_id`: Citation identification number
  - `categ`: Category of the citation

### 13. `TLS216_APPLN_CONT`
- **Fields:**
  - `appln_id`: Application identification number
  - `cont_appln_id`: Continued application ID
  - `cont_type`: Type of continuation

### 14. `TLS219_APPLN_QUAL`
- **Fields:**
  - `appln_id`: Application identification number
  - `appln_qual`: Quality indicator of the application

### 15. `TLS221_INPADOC_FAM`
- **Fields:**
  - `fam_id`: INPADOC family ID
  - `appln_id`: Application identification number

### 16. `TLS223_APPLN_NCLS`
- **Fields:**
  - `appln_id`: Application identification number
  - `ncls_class_symbol`: NCLS classification symbol

### 17. `TLS224_APPLN_CPC`
- **Fields:**
  - `appln_id`: Application identification number
  - `cpc_class_symbol`: CPC classification symbol

### 18. `TLS225_DOCDB_FAM`
- **Fields:**
  - `docdb_family_id`: DOCDB family ID
  - `appln_id`: Application identification number

### 19. `TLS226_DOCDB_FAM_CPC`
- **Fields:**
  - `docdb_family_id`: DOCDB family ID
  - `cpc_class_symbol`: CPC classification symbol

### 20. `TLS227_PERS_PUBLN`
- **Fields:**
  - `person_id`: Person identification number
  - `pat_publn_id`: Patent publication identification number

## Person Tables

### 21. `TLS206_PERSON`
- **Fields:**
  - `person_id`: Person identification number
  - `person_name`: Name of the person
  - `person_address`: Address of the person
  - `person_ctry_code`: Country code of the person

## Event Tables

### 22. `TLS231_INPADOC_LEGAL_EVENT`
- **Fields:**
  - `event_id`: Legal event identification number
  - `appln_id`: Application identification number
  - `event_date`: Date of the legal event
  - `event_code`: Code of the legal event
  - `event_auth`: Authority of the legal event

## Concordance and Technology Field Tables

### 23. `TLS901_TECHN_FELD_IPC`
- **Fields:**
  - `ipc_class_symbol`: IPC classification symbol
  - `techn_field`: Technological field

### 24. `TLS902_IPC_NACE2`
- **Fields:**
  - `ipc_class_symbol`: IPC classification symbol
  - `nace_code`: NACE code
