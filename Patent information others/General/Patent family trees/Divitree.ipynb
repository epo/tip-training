# Conclusion

The divisional tree displayed in the Sunburst chart depicts the hierarchical structure of an extended patent family derived from INPADOC (International Patent Documentation). The data for this visualization was sourced via Open Patent Services (OPS) and pertains to the input family. Each segment represents a patent document, such as European Patent (EP) or World Intellectual Property Organization (WO) applications, with relationships defined by divisional applications and continuations.
* Each colour in the Sunburst chart of the INPADOC family matches a different simple patent family, also called DOCDB family.
* The central node, given in example in that notebook (p. ex. EP09164213, EP99202455 or EP06844225), serves as the root of the family tree, representing the original patent application, while the input patent application is indexed with an '*'.
* The surrounding branches illustrate divisional and associated patent filings, organized by their legal and technical lineage.
* Hierarchical depth indicates the sequence of divisional applications or extensions, demonstrating how the family evolved over time.
* The use of a Sunburst chart provides an intuitive means to explore the intricate connections within this patent family, enabling legal and technical stakeholders to assess its scope and coverage effectively.
* The patents belonging to the same circle in the divisional tree represent parallel or sister applications within the same patent family. These sister divisional applications might be filed close in time or not, however still belonging to the same generation, sharing a common priority date and originating from the same initial application. They generally claim variations of the same invention, often focusing on different aspects of the technology to address specific jurisdictions or legal requirements.
