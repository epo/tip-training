#### 7.1 Advanced searches in bibliographic data, claims, and full-text patent documents
In the **Online Patent Search (OPS)** system provided by the European Patent Office (EPO), you can perform detailed and advanced searches in bibliographic data, claims, and full-text patent documents. Here's how you can structure searches and which operators are available:

### 7.1.1 **Basic Search Structure**
OPS allows users to search patent data by different fields such as title, abstract, claims, inventor, applicant, classification, and full text. The fields used for search can be specified by their respective field identifiers, such as:
- **TI** (title)
- **AB** (abstract)
- **CL** (claims)
- **DE** (description)

To search the full text or a specific part of a patent, the relevant field identifier must be used. For example:
- **CL="autonomous vehicle"** will search for "autonomous vehicle" in the claims section.
- **TI="smartphone"** will search for "smartphone" in the title.

### 7.1.2. **Search Instructions, Operators, and Wildcards**

#### **Boolean Operators**:
- **AND**: Finds documents containing all the search terms.
   - Example: `CL="electric vehicle" AND TI="battery"`
- **OR**: Finds documents containing any of the search terms.
   - Example: `TI="solar panel" OR CL="solar energy"`
- **NOT**: Excludes documents containing a specific term.
   - Example: `TI="drone" NOT CL="military"`

#### **Proximity Operators**:
- **NEAR/n**: Finds terms within a certain distance of each other in the text.
   - Example: `CL="renewable NEAR/5 energy"` searches for "renewable" within five words of "energy" in the claims.
- **WITHIN/n**: Specifies proximity between search terms, though it may depend on specific implementations.
  
#### **Wildcards**:
- **?**: Replaces a single character in a word.
   - Example: `batt?ry` will find "battery" or "battary."
- **\***: Replaces zero or more characters in a word.
   - Example: `automob*` will find "automobile," "automobiles," or "automobilist."

#### **Phrase Searching**:
- Use **quotation marks** (" ") to search for exact phrases.
   - Example: `"artificial intelligence"` will find instances where the exact phrase "artificial intelligence" appears.
  
### 7.1.3. **Truncation and Stemming**
- **Truncation**: The wildcard `*` can be used for truncation. 
   - Example: Searching `robot*` will find "robot," "robots," "robotic," and other variations.
- **Stemming**: The system often supports stemming, so searching `run` may automatically return results for "running," "runner," etc., but using explicit wildcards like `run*` ensures all variations.

### 7.1.4. **Range Searches**:
- You can search for numeric ranges for fields like publication date or application number.
   - Example: `PD=[20150101 TO 20200101]` (for patents published between 1 January 2015 and 1 January 2020).

### 7.1.5. **Classification Searches**:
You can use classification codes like the **International Patent Classification (IPC)** or the **Cooperative Patent Classification (CPC)** to filter searches by technical fields.
   - Example: `IC="G06F17"` will search for patents classified under G06F17 (related to data processing systems).

### 7.1.6. **Citation Searches**:
You can use a cited publication number like EP0351918 to retrieve citing patents.
   - Example: `ct=EP0351918` will search for patents citing EP0351918.

### 7.1.7. **Field-Specific Searches**:
You can search within specific sections of the patent document:
   - **CL** for claims
   - **DE** for description
   - **PA** for applicant
   - **IN** for inventor
   - **PN** for publication number
   - **AN** for application number

For example:
   - `CL="neural network" AND PA="Google"` (searches for patents by Google in the claims mentioning "neural network").

### 7.1.8. **Combining Search Fields**:
You can combine different search fields using Boolean operators:
   - Example: `TI="machine learning" AND IC="G06N"` searches for patents with "machine learning" in the title and classified under G06N (related to artificial intelligence).

### 7.1.9. **Examples**:
- **Claims Search Example**: 
   - Example: `CL="renewable energy" AND AB="solar power"`
   This will find patents where the claims mention "renewable energy" and the abstract mentions "solar power."
   
- **Full Text Search Example**: 
   - Example: `DE="autonomous driving" AND PA="Tesla"`
   This will find patents where the description includes "autonomous driving" and the applicant is Tesla.

### 7.1.10. **Complex Queries**:
You can use parentheses to create more complex queries:
   - Example: `(TI="AI" OR CL="machine learning") AND (IC="G06N" OR IC="G06F")`
   This query finds patents where either the title or claims mention "AI" or "machine learning" and are classified under AI-related categories.

### Conclusion
The OPS search tool is powerful for retrieving patent data based on specific bibliographic fields, claims, and full text. You can refine searches using Boolean operators, proximity operators, wildcards, and classification codes to perform complex patent searches.
