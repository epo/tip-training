# Unitary Patent procedural data: `REG721_PROC_STEP`

The table `REG721_PROC_STEP`contains information regarding procedural data published in the EP register.
Let’s break down the key fields in this table. The first step as usual is to initialise the PATSTAT client and import the table.

```python
from epo.tipdata.patstat import PatstatClient

# Initialise the PATSTAT client
patstat = PatstatClient(env='PROD')

# Access ORM
db = patstat.orm()

# Importing the models
from epo.tipdata.patstat.database.models import REG721_PROC_STEP

# Importing sql alchemy package
from sqlalchemy import func
```

## Key fields in the `REG721_PROC_STEP` table
* `ID` (primary key): a technical identifier for the unitary patent, without business meaning.
* `STEP_ID`: this attribute contains IDs identifying a procedural step.
* `STEP_PHASE`: this attribute indicates the phase during which a procedural step took place.
* `STEP_CODE`: this attribute contains the code assigned to the procedural step.
* `STEP_RESULT`: this attribute contains the procedural step result.
* `STEP_RESULT_TYPE`: this attribute contains the result type of the procedural step.
* `STEP_COUNTRY`: this attribute contains the office code of the state affected by a procedural step. The code identifies the office where the preliminary examination was carried out. Note: currently, only preliminary examinations for international applications have an associated code. This code identifies the office where the preliminary examination was carried out. 
* `TIME_LIMIT`:this attribute contains the time limits associated with the procedural steps. The default value is the empty string.
* `TIME_LIMIT_UNIT`: this attribute specifies the unit of the time limits associated with the procedural steps. The default value is the empty string.
* `BULLETIN_YEAR`: this attribute indicates the year of the publication in the European Patent Bulletin.
* `BULLETIN_NR`: this attribute indicates indicates the calendar week in which the European Patent Bulletin has been published.

Let's plot the table to visualise its structure.

```python
reg721 = (
    db.query(
         REG721_PROC_STEP.id,
         REG721_PROC_STEP.step_id,
         REG721_PROC_STEP.step_phase,
         REG721_PROC_STEP.step_code,
         REG721_PROC_STEP.step_result,
         REG721_PROC_STEP.step_result_type,
         REG721_PROC_STEP.step_country,
         REG721_PROC_STEP.time_limit,
         REG721_PROC_STEP.time_limit_unit,
         REG721_PROC_STEP.bulletin_year,
         REG721_PROC_STEP.bulletin_nr,
    )
    .order_by( REG721_PROC_STEP.id)
    .limit(10000)
)

reg721_df = patstat.df(reg721)
reg721_df
```
