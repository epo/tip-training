# Unitary Patent procedural data: `REG722_PROC_STEP_TEXT`
The `REG722_PROC_STEP_TEXT` table contains a short explanation or more detailed information about each procedural step from the `REG721_PROC_STEP` table, offering additional context and explanations for each step. 
Let’s break down the key fields in this table. The first step as usual is to initialise the PATSTAT client and import the table.

```python
from epo.tipdata.patstat import PatstatClient

# Initialise the PATSTAT client
patstat = PatstatClient(env='PROD')

# Access ORM
db = patstat.orm()

# Importing the as models
from epo.tipdata.patstat.database.models import REG722_PROC_STEP_TEXT,REG721_PROC_STEP

# Importing sql alchemy package
from sqlalchemy import func
```

## Key fields in the `REG722_PROC_STEP_TEXT` table
* `ID` (primary key): a technical identifier for the Unitary Patent, without business meaning.
* `STEP_ID`: this attribute contains IDs identifying a procedural step.
* `STEP_TEXT`: this attribute contains a text related to the procedural step. It provides a description or explanation related to the specific procedural step identified in the `REG721_PROC_STEP table`. This text offers additional details or context about the step, helping users better understand its purpose and associated actions.
* `STEP_TEXT_TYPE`: this attribute specifies the type of text associated with the procedural step. It categorizes the text according to its content or purpose, such as a description of the step, references to specific years, or details about extension states. The text type helps users understand the context in which the text is used.

Let's plot the table to visualize its structure.

```python
reg722 = (
    db.query(
         REG722_PROC_STEP_TEXT.id,
         REG722_PROC_STEP_TEXT.step_id,
         REG722_PROC_STEP_TEXT.step_text,
         REG722_PROC_STEP_TEXT.step_text_type,
    )
    .order_by( REG722_PROC_STEP_TEXT.id)
    .limit(10000)
)

reg722_df = patstat.df(reg722)
reg722_df
```

Let’s join the `REG722_PROC_STEP_TEXT` table with the `REG721_PROC_STEP` table to obtain a more complete overview of the information related to each procedural step.

```python
reg722 = (
    db.query(
         REG722_PROC_STEP_TEXT.id,
         #REG722_PROC_STEP_TEXT.step_id,
         REG722_PROC_STEP_TEXT.step_text,
         REG722_PROC_STEP_TEXT.step_text_type,
         REG721_PROC_STEP.id,
         REG721_PROC_STEP.step_id,
         REG721_PROC_STEP.step_phase,
         REG721_PROC_STEP.step_code,
         REG721_PROC_STEP.step_result,
         REG721_PROC_STEP.step_result_type,
    ).join(
         REG722_PROC_STEP_TEXT,
        REG722_PROC_STEP_TEXT.id== REG721_PROC_STEP.id
            
    ).order_by(REG721_PROC_STEP.id)
    .limit(10000)
)

reg722_df = patstat.df(reg722)
reg722_df
        
        
        
```
