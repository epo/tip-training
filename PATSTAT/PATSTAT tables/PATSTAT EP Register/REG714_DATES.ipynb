# Unitary patents dates: `REG714_DATES`

The `REG714_DATES` table contains several dates associated with Unitary Patents.
Let’s break down the key fields in this table. The first step as usual is to initialise the PATSTAT client and import the table.

```python
from epo.tipdata.patstat import PatstatClient

# Initialise the PATSTAT client
patstat = PatstatClient(env='PROD')

# Access ORM
db = patstat.orm()

# Importing the as models
from epo.tipdata.patstat.database.models import REG714_DATES

# Importing sql alchemy package
from sqlalchemy import func
```

## Key fields in the `REG714_DATES` table
* `ID` (primary key): a technical identifier for the Unitary Patent, without business meaning.
* `BULLETIN_YEAR`: this attribute indicates the year of the publication in the European Patent Bulletin.
* `BULLETIN_NR`: this attribute indicates indicates the calendar week in which the European Patent Bulletin has been published.
* `DATE_TYPE`: this attribute defines the type of date. There are seven possible values:
    * **RQEXAM**: request for examination — the date on which the request for examination was validly filed.
    * **FEXRRQ**: first examination report requested — the date of dispatch of the first communication from the examining division.
    * **PMNTAM**: patent maintained as amended — the publication date of the B2 document (patent maintained as amended).
    * **PRCITR**: proceedings interrupted — the date on which the procedure was interrupted, together with the cause of the interruption.
    * **PRCRES**: proceedings resumed — the date on which the procedure was resumed.
    * **PRCSUS**: proceedings suspended — the date on which the procedure was suspended, together with the reason for the suspension.
    * **RQCONV**: request for conversion to a national application.
* `EVENT_DATE`: the date corresponding to the `DATE_TYPE` attribute
* `CAUSE_INTERRUPTION`: this attribute specifies the reasons for the interruption of the proceedings, the occurences are:
    * **NA**: not applicable or unknown
    * **DA**: death of applicant
    * **DR**: death of representative
    * **IA**: legal incapacity of applicant
    * **IR**: legal incapacity of representative
    * **PA**: legal prevention of applicant
    * **PR**: legal prevention of representative
* `CONVERTED_TO_COUNTRY`: this attribute contains the office (country) to which the application was converted into a national application.
  
Let's plot the table to visualise its structure.

```python
reg714 = (
    db.query(
         REG714_DATES.id,
         REG714_DATES.bulletin_year,
         REG714_DATES.bulletin_nr,
         REG714_DATES.date_type,
         REG714_DATES.event_date,
         REG714_DATES.cause_interruption,
         REG714_DATES.converted_to_country,
    )
    .order_by( REG714_DATES.id)
    .limit(10000)
)

reg714_df = patstat.df(reg714)
reg714_df
```
