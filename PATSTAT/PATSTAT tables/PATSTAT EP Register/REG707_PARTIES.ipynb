# Proprietor, inventor, legal representative (`REG707_PARTIES`)
The table `REG707_PARTIES`contains proprietors, inventors and legal representatives of Unitary Patents.
Let’s break down the key fields in this table. The first step as usual is to initialise the PATSTAT client and import the table.

```python
from epo.tipdata.patstat import PatstatClient

# Initialise the PATSTAT client
patstat = PatstatClient(env='PROD')

# Access ORM
db = patstat.orm()

# Importing the models
from epo.tipdata.patstat.database.models import REG707_PARTIES

# Importing sql alchemy package
from sqlalchemy import func
```

## Key fields in the `REG707_PARTIES` table
* `ID` (primary key): a technical identifier for the Unitary Patent, without business meaning.
* `SET_SEQ_NR`: For a given patent (identified by its ID in the table), the value of SET_SEQ_NR identifies which temporal version of the set of applicants, inventors, or representatives is being considered. For instance, if we have a Unitary Patent for which the original applicants are A1 and A2, the set containing A1 and A2 will be identified by `SET_SEQ_NR` with number 1 (oldest version). If later, for various reasons, the set of applicants changes (for instance, A1 remains the only applicant) `SET_SEQ_NR` will change accordingly (SET_SEQ_NR will assume value 2). The numbering is assigned in descending order within each party set (applicants, inventors, or representatives).
* `IS_LATEST`: indicates the latest set of applicants / inventors / representatives / opponents. It is an Y/N flag. It assume value Y if the set is the latest (= current or most recent), it assume value N otherwise.
* `CHANGE_DATE`: this attribute indicates the date when the data was modified in the corresponding application of the EPO. It represents the date on which the record was saved in the database.
* `BULLETIN_YEAR`: this attribute indicates the year of the publication in the European Patent Bulletin.
* `BULLETIN_NR`: this attribute indicates indicates the calendar week in which the European Patent Bulletin has been published.
* `SEQ_NR`: this attribute contains the sequence number of party (concerns applicant, inventor and legal representative).
* `ADDRESS_1`, `ADDRESS_2`,`ADDRESS_3`, `ADDRESS_4`, `ADDRESS_5`: these attributes contain the first / second / third / forth / fifth address line of a person. Address lines from the second onward are often empty. 
* `COUNTRY`: this attribute contains the country/territory code. It is a two-letter country/territory code for patent parties (applicant/inventor/agent), designated states of applicant, or country of licensees.
* `TYPE`: this attribute indicates the type of party ("A" for applicant / "I" for inventor / "R" for legal representative).
* `WISHES_TO_BE_PUBLISHED`: this attribute indicates whether an inventor has waived the right to be designated. When an inventor waives this entitlement, the EP Register does not publish the inventor’s name or address. Nevertheless, the existence of an inventor is explicitly recorded, even though the name and address are not disclosed. If the attribute `WISHES_TO_BE_PUBLISHED` is present and set to “no”, the attribute assumes value ‘N’. Otherwise, the default value (empty string) is applied.
* `DESIGNATION`: this attribute indicates the type of designation. 
* `CUSTOMER_ID`: this attribute contains the identifiers of customers, who may be legal or natural persons. If a customer holds multiple roles (applicant, inventor, representative, licensee, opponent), they usually have separate CUSTOMER_IDs for each role.
* `NAME`: This attribute contains the names of natural and legal persons.

Let's plot the table to visualise its structure.

```python
reg707 = (
    db.query(
        REG707_PARTIES.id,
        REG707_PARTIES.set_seq_nr,
        REG707_PARTIES.is_latest,
        REG707_PARTIES.bulletin_year,
        REG707_PARTIES.bulletin_nr,
        REG707_PARTIES.seq_nr,
        REG707_PARTIES.address_1,
        REG707_PARTIES.address_2,
        REG707_PARTIES.address_3,
        REG707_PARTIES.country,
    )
    .order_by(REG707_PARTIES.id)
    .limit(10000)
)

reg707_df = patstat.df(reg707)
reg707_df
```
